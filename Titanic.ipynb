```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.info()
```

# 성별에 따른 승객 수

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
sns.countplot(data=titanic, x="sex", hue="sex", palette="pastel", edgecolor="black", legend=False)
plt.title("Passenger Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

# 객실 등급 별 승객 수

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

sns.countplot(data=titanic, x="class", palette="Set2", edgecolor="black")

plt.title("Number of Passengers by Class", fontsize=14)
plt.xlabel("Class", fontsize=12)
plt.ylabel("Passenger Count", fontsize=12)

plt.tight_layout()
plt.show()
```

# 선실 층 별 승객 수

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

sns.countplot(data=titanic, x="deck", palette="Set3", edgecolor="black")

plt.title("Passenger Count by Deck", fontsize=14)
plt.xlabel("Deck", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.tight_layout()
plt.show()
```

# 객실 등급에 따른 생존 여부 별 승객 수

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

sns.countplot(data=titanic, x='class', hue='alive', palette='Set2', edgecolor='black')

plt.title("Survival Count by Passenger Class", fontsize=14)
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Survived")
plt.tight_layout()
plt.show()
```

# 성별에 따른 평균 생존률

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

plt.figure(figsize=(6, 5))
ax = sns.barplot(data=titanic, x="sex", y="survived", hue="sex", palette="pastel", edgecolor="black", legend=False)

plt.title("Average Survival Rate by Sex", fontsize=16, weight="bold")
plt.xlabel("Sex", fontsize=13)
plt.ylabel("Survival Rate", fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

# 객실 등급과 성별에 따른 평균 탑승 비용 비교

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.figure(figsize=(8,6))

ax = sns.barplot(data=titanic, x="class", y="fare", hue="sex", palette="Set2", edgecolor="black")

plt.title("Average Fare by Passenger Class and Sex", fontsize=16, weight="bold")
plt.xlabel("Passenger Class", fontsize=14)
plt.ylabel("Average Fare", fontsize=14)

plt.legend(title="Sex", fontsize=12, title_fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

# 선실 층별 객실 등급에 따른 평균 탑승 요금 비교

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.figure(figsize=(10,6))

ax = sns.barplot(data=titanic, x="deck", y="fare", hue="class", palette="Set3", edgecolor="black")

plt.title("Average Fare by Deck and Passenger Class", fontsize=16, weight="bold")
plt.xlabel("Deck", fontsize=14)
plt.ylabel("Average Fare", fontsize=14)

plt.legend(title="Passenger Class", fontsize=12, title_fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

# 선실 층과 객실 등급에 따른 평균 생존률 비교 그래프

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.figure(figsize=(10,6))

ax = sns.barplot(data=titanic, x="deck", y="survived", hue="class", palette="Set2", edgecolor="black")

plt.title("Average Survival Rate by Deck and Passenger Class", fontsize=16, weight="bold")
plt.xlabel("Deck", fontsize=14)
plt.ylabel("Average Survival Rate", fontsize=14)

plt.legend(title="Passenger Class", fontsize=12, title_fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```
